```python
# Resources require for script

from bs4 import BeautifulSoup
import requests
import pandas as pd


url = "https://en.wikipedia.org/wiki/List_of_Asian_stock_exchanges"

r = requests.get(url)
df_list = pd.read_html(r.text) # this parses all the tables in webpages to a list
```

```python
len(df_list)
```

```python
data = r.text
soup = BeautifulSoup(data,'html.parser')
headings = list()
for header in soup.find_all('h3'): ## find table names
  
  if (not (header.next_element.find('span'))):
    headings.append(header.next_element)
```

```python
for i in range(len(headings)):
  print("{}:{}".format(i,headings[i].text))
```
