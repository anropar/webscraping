```python
# Resources require for script

from bs4 import BeautifulSoup
import requests
import pandas as pd


url = "https://en.wikipedia.org/wiki/List_of_Asian_stock_exchanges"

r = requests.get(url)
df_list = pd.read_html(r.text) # this parses all the tables in webpages to a list
```

```python
len(df_list)
```

```python
st = {0:'Central Asian',1:'East Asian',2:'North Asia',3:'South Asian',4:'Southeast Asian',5:'West Asian'}
```

```python

stock_exchanges = list()
for st_e in df_list:
  stock_exchanges.append(st_e)
```
